## Filter A DataFrame Based On A Condition

```python
df = pd.read_csv("https://raw.githubusercontent.com/pujaroy280/ML-Data/main/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
```

```python
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

## Filter with More than One Condition (AND)

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"
```

```python
df[mask1 & mask2]
```

## Filter with More than One Condition (OR)

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

## The .isnull() and .notnull() Methods

```python
mask = df["Team"].isnull()

df[mask]
```

```python
condition = df["Gender"].notnull()

df[condition]
```

## The .between() Method

```python
df[df["Salary"].between(60000, 70000)]

'''
Return boolean Series equivalent to left <= series <= right.

This function returns a boolean vector containing True wherever the corresponding Series element is between the boundary values left and right. NA values are treated as False.

'''
```

```python
df[df["Bonus %"].between(2.0, 5.0)]
```

```python
df[df["Start Date"].between("1991-01-01", "1992-01-01")]
```

```python
df[df["Last Login Time"].between("08:30AM", "12:00PM")]
```

## The .duplicated() Method

```python
df.sort_values("First Name", inplace = True)
df.head()
```

```python
mask = df["First Name"].duplicated(keep = False)
df[mask]
```

```python
mask = ~df["First Name"].duplicated(keep = False)
df[mask]

'''
~n operator is the bitwise negation operator: it takes the number n as binary number 
and “flips” all bits 0 to 1 and 1 to 0 to obtain the complement binary number. For example, the tilde operation ~1 becomes 0 and ~0 becomes 1 and ~101 becomes 010
'''
```
